## News crawler

#### TVBS

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 開啟文件檔讀取網址
```

```python
url = 'https://news.tvbs.com.tw/life/1569771'
response = requests.get(url)#headers=headers)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
tags = soup.find_all('div',{'class':'article_content'})
```

```python
for tag in tags:
    gg = tag.get_text()
# gg2 = gg.split('。')
# gg.pop()
dd = []
for cc in gg:
    dd.append(cc)
dd
```
